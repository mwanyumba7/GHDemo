# Advanced Excel Features

This section introduces **advanced Excel tools and functions** that make data analysis faster and more efficient.  
These features are essential for handling **large datasets**, performing **complex lookups**, and **transforming data**.

## 1. LOOKUP Functions

LOOKUP functions help you **search for a value** in a range or table and return a matching result.  
The most common lookup functions are **VLOOKUP**, **HLOOKUP**, and **XLOOKUP**.

### a) VLOOKUP (Vertical Lookup)
Searches for a value in the **first column** of a table and returns a value in the same row from a specified column.

**Syntax:**

`=VLOOKUP(lookup_value, table_array, col_index_num, [range_lookup])`

`Note:` The column which holds the data used to lookup must always be to the left.

- `Lookup_value`: Select the cell where search values will be entered.
- `Table_array`: The table range, including all cells in the table.
- `Col_index_num`: The data which is being looked up. The input is the number of the column, counted from the left:
- `Range_lookup`: TRUE if numbers (1) or FALSE if text (0).

![VLOOKUP](https://github.com/user-attachments/assets/17d74099-86b3-478c-b203-7c743d9f3ec0)

### b) HLOOKUP (Horizontal Lookup)
Works like VLOOKUP but searches for a value in the **first row** instead of the first column.

![HLOOKUP](https://github.com/user-attachments/assets/ed8d172c-7bf9-47e1-affd-791b6e24c1e1)

### c) XLOOKUP (Newer Excel Versions)
A more flexible replacement for VLOOKUP and HLOOKUP.  
It can search both **horizontally and vertically**, and you don’t need to specify column numbers.

**Syntax:**
`=XLOOKUP(lookup_value, lookup_array, return_array, [if_not_found])`

**Example:**
`=XLOOKUP(101, A2:A20, D2:D20, "Not Found")`
Finds ID `101` in `A2:A20` and returns the matching value from `D2:D20`.

## 2. INDEX and MATCH

**Video Resource:** [Excel Index and Match](https://youtu.be/SW075RZ5188?si=IDIj5JVSiFByX73c)

`INDEX` and `MATCH` are often used together as a **more powerful alternative** to VLOOKUP.

### INDEX
Returns the value of a cell within a range based on row and column numbers.

**Syntax:**
`=INDEX(array, row_num, [column_num])`

### MATCH
Returns the **position** of a value in a range.

**Syntax:**
`=MATCH(lookup_value, lookup_array, [match_type])`


### Using INDEX + MATCH
Combine both to look up a value anywhere in a table:

`=INDEX(D2:D20, MATCH(101, A2:A20, 0))`

- Finds the position of `101` in column `A2:A20` using MATCH.
- Uses INDEX to return the corresponding value from column `D2:D20`.

**Advantages over VLOOKUP:**
- Works even if the lookup column is not the first column.
- Faster for large datasets.

## 3. Power Query (Get & Transform)

Power Query is a tool for **importing**, **cleaning**, and **transforming** data from different sources  
(e.g., Excel files, CSV, databases, web data).

With Power Query (known as Get & Transform in Excel), you can import or connect to external data, and then shape that data, for example remove a column, change a data type, or merge tables, in ways that meet your needs. 

Then, you can load your query into Excel to create charts and reports. Periodically, you can refresh the data to make it up to date. 

**Key Features:**
- Remove duplicates.
- Split columns.
- Combine multiple tables.
- Automate repetitive cleaning tasks.

```excel
=TEXT(value, format_text)

## Introduction to Pivot Tables

A `Pivot Table` is a powerful Excel feature used to summarize, analyze, explore, and present large datasets.

It allows you to reorganize and group data (without changing the original data) to generate quick summaries such as totals, averages, counts, or percentages.

### Pivot Table Components

When you create a Pivot Table, Excel provides four main areas to place fields:

| Area        | Purpose                                  | Example                     |
| ----------- | ---------------------------------------- | --------------------------- |
| **Rows**    | Defines how data is grouped vertically.  | Products, Regions           |
| **Columns** | Defines grouping horizontally.           | Months, Years               |
| **Values**  | Displays the summarized calculations.    | Total Sales, Average Profit |
| **Filters** | Allows filtering the entire Pivot Table. | Year = 2024                 |

### Steps to Create a Pivot Table

#### Step 1: Preparing the Data

Before creating a pivot table, ensure our data is properly formatted:

- **Organize in a Tabular Format**: Place our data in rows and columns, with each column having a header.
- **Avoid Blank Rows or Columns**: Ensure there are no empty rows or columns within our dataset.
- **Name your Data Range (Optional)**: Highlight your data and assign a name with Formulas > Define Name for easier reference.



#### Step 2: Selecting the Data

1. Click any cell inside our data or
2. Highlight the specific range we want to include in the pivot table.

#### Step 3: Inserting a Pivot table

- Go to the Insert tab on the Excel ribbon.
- Click PivotTable.
- In the Create PivotTable dialog box:
- Verify the selected data range.
- Choose the location:
- New Worksheet: Places the Pivot Table in a new sheet (recommended).
- Existing Worksheet: Specify a cell in the current sheet.



#### Shortcut Keys

- **Windows**: Press Alt + N + V to open the Create PivotTable dialog box.
- **Mac**: Press Command + Option + P to create a pivot table.



#### Step 4: Build your Pivot Table

You'll see a PivotTable Field List pane on the right side of your screen. This is where you organize your data:



##### a) Drag and Drop Fields:

Drag column headers from the Field List into one of the four areas:

- **Rows**: Sets rows for the table.
- **Columns**: Creates columns for our data.
- **Values**: Adds numerical data to be calculated like sum, count, etc.
- **Filters**: Adds filters to refine our analysis.



##### b) Customize Calculations:

Right-click on a value in the Values area and choose Value Field Settings. Then, Select the desired calculation like Sum, Average, Count, etc.

#### Step 5: Formatting and Customizing the Pivot Table

- **Apply a PivotTable Style**: Select the pivot table and go to Design > PivotTable Styles to apply a pre-designed format.
- **Sort and Filter**: Use the dropdown arrows on row or column headers to sort and filter data.
- **Group Data**: Right-click on a row or column item and select Group to organize data by date, number ranges etc.
- **Add Slicers (Optional)**: Go to Insert > Slicer to create interactive filters for our pivot table.

#### Step 6: Refresh the Pivot Table

Update the pivot table when source data changes.Clicking anywhere in the pivot table.
- Going to **PivotTable Analyze > Refresh**. 

##### Select Entire Pivot Table Shortcut Key
**Windows/Mac**: Press Ctrl + A (or Command + A on Mac) to select the entire pivot table.
